```python
#6. in sql task try to perform left join operation with attrubute dataset and dress dataset on column Dress_ID

# Importing connector
import mysql.connector as connection
import pandas as pd
import sqlalchemy
#Establishing Connection
mydb = connection.connect(host = 'localhost', user = 'root', password = 'thanmai123')
engine = sqlalchemy.create_engine('mysql+pymysql://root:thanmai123@localhost:3306/assignment')
dress_sales = pd.read_sql_table('dress_sales',engine)

query = '''
 SELECT *
 FROM Attribute_dataset LEFT JOIN dress_sales
 ON Attribute_dataset.Dress_ID=dress_sales.Dress_ID
'''
df = pd.read_sql_query(query,engine)
df
```

```python
#7. Write a sql query to find out how many unique dress that we have based on dress id

mydb = connection.connect(host = 'localhost', user = 'root',database = 'Assignment', password = 'thanmai123')

q6 = '''
 SELECT count(distinct(dress_sales.Dress_ID))
 FROM Attribute_dataset LEFT JOIN dress_sales
 ON Attribute_dataset.Dress_ID=dress_sales.Dress_ID
'''
cursor = mydb.cursor() 
cursor.execute(q6)
print(cursor.fetchall())
```

```python
#8. Try to find out how mnay dress is having recommendation 0

q7 = '''
select count(recommendation)
from attribute_dataset
where recommendation = 0
'''

cursor = mydb.cursor() 
cursor.execute(q7)
print(cursor.fetchall())
```

```python
#9. Try to find out total dress sell for individual dress id
q8 = '''
 select Dress_ID,sum(`29/8/2013`+`31/8/2013`+`09-02-2013`+`09-04-2013`+`09-06-2013`+`09-08-2013`+ `09-10-2013`+ `09-12-2013`+
  `14/9/2013`+ `16/9/2013`+ `18/9/2013`+ `20/9/2013`+ `22/9/2013`+ `24/9/2013` + `26/9/2013`+ `28/9/2013`+ `30/9/2013`+ 
  `10-02-2013`+`10-04-2013`+`10-06-2013`+`10-08-2010`+`10-10-2013`+`10-12-2013`) Total_sales from dress_sales
group by Dress_ID
order by Total_sales desc;
'''

pd.read_sql_query(q8,engine)
```

```python
#10. Try to find out a third highest most selling dress id 
q9 = '''
 select Dress_ID,sum(`29/8/2013`+`31/8/2013`+`09-02-2013`+`09-04-2013`+`09-06-2013`+`09-08-2013`+ `09-10-2013`+ `09-12-2013`+
  `14/9/2013`+ `16/9/2013`+ `18/9/2013`+ `20/9/2013`+ `22/9/2013`+ `24/9/2013` + `26/9/2013`+ `28/9/2013`+ `30/9/2013`+ 
  `10-02-2013`+`10-04-2013`+`10-06-2013`+`10-08-2010`+`10-10-2013`+`10-12-2013`) Total_sales from dress_sales
group by Dress_ID
order by Total_sales desc limit 1 offset 2;
'''

pd.read_sql_query(q9,engine)
```
